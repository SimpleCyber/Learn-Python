```python
# reading entire using readline

f = open('multiline.txt','r')
satyam =True
while satyam:
    data = f.readline() 
    # readline while it is not empty
    if data == ' ':
        satyam = False
        break
    else:
        print(data ,end =' ')

f.close()
```

Using Context Manager (WITH)

```python
# it closes a file when it use is over 

# with =>WRITE

with open('sample2.txt','w') as f:
    f.write('selomon satyam bro')
```

```python
# with =>READ

with open('sample.txt','r') as f:
    print(f.read())
```

```python
with open('sample.txt','r') as f:
    print(f.read(10))
    print(f.read(10))
```

```python
big_data = ['hello world ' for i in range(1000)]

with open('sig_f.txt','w') as f:
    f.writelines(big_data)
```

```python
with open('sig_f.txt','r') as f:
    chunk_size = 100
    while len(f.read(chunk_size)) > 0:
        print(f.read(chunk_size),end="")
```

```python
# SEEK and TELL function

with open('sample.txt','r') as f:
    print(f.read(10))

    print(f.tell()) # number on which next going to print  abhi tak aap ne kitna text likha hai


    f.seek(0) # jo print ka value 10 pe tha wapas 0 se start karo print karna 
    print(f.read(10))

    print(f.tell())
```

```python
# SEEK during Write

with open('sample.txt','w') as f:
    f.write('Hello')
    f.seek(0)
    f.write('X')
    f.write('Xa')


    # X will replace H 
```

Problems related to text file 
 - can not work with binary files like videos and image s
 - not good for other data types like [int/float/list/tuples]

```python
# working with binary files 

with open('boy.png','r') as f:
    f.read()
```

```python
# working with binary files
# rb => read binary ------------------
with open('boy.png','rb') as f:
    with open('boy_copy.png','wb') as wf:
        wf.write(f.read())
```

```python
# not working with other data types

with open('sample.txt','w') as f:
    f.write(5)
    # only textual values can be written
```

```python
with open('sample.txt','w') as f:
    f.write("5")
```

```python
# more COMPLEX data

d ={
    'name' : 'satyam',
    'age ': 19,
    'gender' : ',male'

}

with open('sample.txt','w') as f:
    f.write(d)
```
