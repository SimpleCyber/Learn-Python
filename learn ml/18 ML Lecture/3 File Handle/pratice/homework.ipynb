```python
# 1) function takes input of file name 
#    returns the final line of the file
x ="""hello sataym
how are u
i hope u are finme
bye 
khuch aur bolna hai kay"""

import json
with open('sample1.json','w') as f:
    json.dump(x,f)

def get_file_line(filename):
    with open(filename,'r') as f:
        return json.load(f)
    
x = get_file_line("sample1.txt")
print(x)
```

```python
# count a,e,i,o,u in a file print using dictionary

with open('sample1.json','r') as f:
    x = json.load(f)


print(x)

vovel ={'a':0,'e':0,'i':0,'o':0,'u':0}

for i in x:

    match i:
        case 'a':
            vovel['a'] +=1
        case 'e':
            vovel['a'] +=1
        case 'i':
            vovel['i'] +=1
        case 'o':
            vovel['o'] +=1
        case 'u':
            vovel['a'] +=1
        
print(vovel)
```
