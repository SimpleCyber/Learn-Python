```python
# serilization using json module
# list 
import json

l= [1,2,3,4,5,6,7,8,9,0]
with open('demo.json','w') as f:
    json.dump(l, f)
    # l => list , f => file handler object
```

```python
# dictionary

d ={
    'name' : 'satyam',
    'age ': 19,
    'gender' : ',male'

}

with open('demo.json','w')  as f:
    json.dump(d , f, indent= 4)
    # indent  for proper display 
```

```python
# DESERILAISATION

import json

with open('demo.json','r') as f:
    d = json.load(f)
    print(d)
    print(type(d))
```

[OBJECTS_of_class]

```python
class Person:
    def __init__(self ,name , lname , age , job):
        self.name = name 
        self.lname  = lname 
        self.age = age 
        self.job = job
```

```python
person = Person('Satyam','Yadav','19','Data Scientist')
```

```python
# my output be like this , 
# Satyam Yadav of age 19 is world greatst Data scientist

# pyhton does not know the format our created object , there fore it will throw an error , so we need to specify the format of our object
```

```python
# As a string

import json
def show_object(person):
    if isinstance (person, Person):
    # isinstance is a way to check the object is of that class or not
        return "{} {} of age {} is world greatst {}".format(person.name ,person.lname,person.age,person.job)
    

with open('demo.json','w') as f:
    json.dump(person,f,default=show_object)
    # UNIQUE WAY 
```

```python
# printing in form of dictionary
import json
def show_object(person):
    if isinstance (person, Person):
    # isinstance is a way to check the object is of that class or not
        return {'name ':person.name + ' ' + person.lname,'age ' :person.age,'job':person.job}

with open('demo.json','w') as f:
    json.dump(person,f,default=show_object, indent=4)
```

```python
# deserilaistion

import json
with open ('demo.json','r') as f:
    d  = json.load(f)
    print(d)
    print(type(d))

    
```
