```python
# ITERATION
def multiply(a,b):
    result = 0
    for i in range (0, b):
        result = a + result

    print(result)

multiply(3,4)
```

```python
# RECURSION

def  multiply(a,b):
    if b == 1:
        return a
    else:
        return a + multiply(a , b-1)
    

print(multiply(3,4))
```

```python
# factorial of a number using recursion

def factorial(num):
    if num == 1:
        return 1
    else :
        return num * factorial(num - 1)
    
print(factorial(5))
```

```python
# palindrome using recursion

# base condition  => jiska answer ham pata hai, in PALINDROME case , if string length is ==1 , then the given string is a palindrome

# else , 1 and last , letter should be same 

string = input("Enter string to check ist id palindrome or not :")

def palinderome(text) :
    if len(text) <= 1:
        print("palinderome")

    else:
        if text[0] == text[-1]:
            return palinderome(text[1:-1])
        else:
            print('Not a palindeome')


palinderome(string)



```

```python
# # THE RABBIT PROBLEM

# - if 2 newborn rabbits are put in a pen , how many rabits will be in the pen after 1 year

# # {conditons}
#     - always reproduce 1 male and female offspring
#     - can reproduce once in a month 
#     - can reproduce once they are 1 month old
#     - never die  


# 0 day => 0 rabbits
# 31 day (1'st month) => 1 pair
# 2'nd mont => 2 pairs
# 3'rd => 3 pairs , oldest new 1 pair (2),prev month born pair (1)  => FIB0NACCI SERIRS

def fibbo(mon):
    if mon == 0 or mon == 1:
        return 1
    else :
        return fibbo(mon - 1) + fibbo(mon -2)
    

print(fibbo(12))


# non linerar behiviour , uppar niche 
```

```python
# DP => MIMEMOSIATIO , time ko reduce karne ke liye , space badha dete hai

def memeo(m,d):
    if m in d:
        return d[m]
    else :
        d[m] = memeo(m-1,d) + memeo(m-2, d)
        return d[m]

d = {0:1 , 1:1}

print(memeo(50, d))

print(d)
```
