```python
# HOF
# higher order of function

# a function takes a input as function
```

```python
def sum (fun , l):
    result = 0
    for i in l:
        if fun(i):
            result = result + i
    return result


l = [11,2,12,13,14,15,16,17,1,8,19,20]


a = lambda x: x%2 == 0 
b = lambda x: x%2 != 0
c = lambda x: x%3 == 0

print(sum(a,l))
```

```python
# MAP
# it always needed a function and an iterratable as an input 
```

```python
#) case 1 , double all values

l = [11,2,12,13,14,15,16,17,1,8,19,20]

list(map(lambda x:x*2, l))
```

```python
# case 2) even odd
list(map(lambda x: x%2 ==0 , l))
```

```python
# case 3) 
dict = [
    {
        'name' :'satyam',
        'roll no':21,
        'job' : 'Data scientist'
    },{
        'name':'rolli',
        'roll no':14,
        'job':'student'
    },{
        'name':'ajay',
        'roll no':96,
        'job': 'master'
    }
]

# usiing map, only print the name in a list

map(lambda y: y['name'], dict)

list(map(lambda y: y['name'], dict))
```

```python
b = ['apple','orange','mango','watermelon']
#  print the string having 'e'

list(filter(lambda x: 'e' in x, b))
```
