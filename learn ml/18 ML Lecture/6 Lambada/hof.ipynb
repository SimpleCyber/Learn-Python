# HOF

```python
# higher order of function

# a function takes a input as function
```

```python
def sum (fun ,l):
    result = 0
    for i in l:
        if fun(i):
            result = result + i
    return result



l = [11,14,21,23,56,78,45,29,28]

a = lambda x: x%2 == 0
b = lambda x: x%2 != 0
c = lambda x: x%3 == 0

print(sum(a,l))
print(sum(b,l))
print(sum(c,l))
```

# MAP

```python
# MAP 

# it always needed a function and an iterratable as an input 
```

```python
#) case 1 , double all values
l = [1,2,3,4,5,6,7,8,9]

map(lambda x: x*2 , l)
```

```python
list(map(lambda x: x*2 , l))
# we need to typecast , so we can read it
```

```python
# case 2) even odd
list(map(lambda x: x%2 == 0,l))
```

```python
# case 3) 
dict = [
    {
        'name' :'satyam',
        'roll no':21,
        'job' : 'Data scientist'
    },{
        'name':'rolli',
        'roll no':14,
        'job':'student'
    },{
        'name':'ajay',
        'roll no':96,
        'job': 'master'
    }
]

# using map , only print the name in a list

map(lambda y: y['name'] , dict)
list(map(lambda y: y['name'] , dict))
```

# REDUCE

```python
from functools import reduce
print(l)
# sum of all
reduce(lambda x,y: x+y ,l )
```

```python
# greatest
reduce(lambda x,y : x if x>=y else y ,l)
```

```python
# smallest
reduce(lambda x,y : x if x<=y else y ,l)
```
