```python
# Lambda are functions

# 1- no name 
# 2- can not reuse
# 3- whole function is returened , insted of returing values
# 4- onle written in 1 line 
```

```python
a = lambda x : x**2
a(4)
```

```python
a = lambda x,y :x*y 
a(2,3)
```

```python
a = lambda x : "Even" if x%2==0 else "Odd"
a(6)
a(1)
```

```python
a = lambda x:  x[0]== 'a'
a('aaple')
```
