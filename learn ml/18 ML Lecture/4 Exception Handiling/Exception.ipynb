```python
# dealing with external file is a risky task always

try:
    with open('sample1.txt','r') as f:
        print(f.read())
except:
    print("File not found ðŸ¤”") 
```

```python
# catching the specific error

try :
    print(m)
except Exception as e:
    #### SARKARI CODDER
    print(e) 
    print(e.with_traceback) # through these we can get technical error printed => {NameError}
```

```python
# catching multiple errors

try :
    m = 5
    print(m)
    print(m/1)
    f =open('sample.txt','r')
    print(f.read())
    l = [1,2,3,4,5,6,7,8]
    l[1000]
except NameError:
    print("Element not declared but used")
except ZeroDivisionError:
    print("Divison by zero is invalid")
except FileNotFoundError:
    print("File not found")

except Exception as e:
    # it always should at last

    # Generic error , any error not declared by us then this will beb executed , as these error is not anticipated by us
    print(e)

```

```python
# ELSE concept 


try :
    f = open('sample.txt','r')
except FileNotFoundError:
    print("File nahi milli")
except Exception:
    print("Kuch to gadbad hai daya")

# jab { TRY } mai koi problem nahi hogi to else execute hoga
else :
    print(f.read())
```

```python
# FINALLY 

# ELSE concept 


try :
    f = open('sample.txt','r')
except FileNotFoundError:
    print("File nahi milli")
except Exception:
    print("Kuch to gadbad hai daya")
else :
    print(f.read())
finally:
    # code bich mai aatak gaya hai to finally ami stop karte hai
    # stop database , bluetoot , network block , etc
    print("Yea to open ho ga hi koi aur option nahi hai")
```
