```python
# rename
import pyttsx3

def speak(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

def func():
    print("oi oi oi ")
    speak("oi oi oi")

a = func # renamed
a()

del func # delte
```

```python
# function as an input 


def modify (func,num):
    return func(num)


def square (num):
    return num **2

modify(square, 5)
```

```python
# DECORATOR 

def outer(func):
    def wrapper():
        print("*********************************")
        func()
        print("*********************************")
    return wrapper

def display():
    print("Hello ðŸ’š")

a = outer(display)
a()
```

```python
# a decorator a , that can record time for any function execution

import time
def timer(func):
    def wraper(*args):
        start = time.time()
        func(*args)
        print("Time taken for execution of", func.__name__, 'is :',time.time() - start)
    return wraper

@timer
def hello():
    print("hello duniya")
    time.sleep(2)


@timer
def power(a,b):
    print( a**b)
    time.sleep(1)

hello()

power(2,3)

```

```python
# DECORATOR PART 2 
# if the correct data type then only execute the condition


def sanity_check(data_type):
    def outer_wraper(func):
        def inner_wrapper(*args):
            # args[0] =>kiu ? , kiu ki yea tupple hai aur first element of a tupple ko fetch kar rahe hai hammm
            if type(args[0]) == data_type:
                func(*args)
            else:
                raise TypeError("Yea data type nahi celaga ðŸ¤”")
        return inner_wrapper
    return outer_wraper
    
    

@sanity_check(int)
def square(num):
    print(num**2)

square(4)
# square("oi oi oi")


@sanity_check(str)
def greet(name):
    print("hello",name)

greet('Satyam')
```
