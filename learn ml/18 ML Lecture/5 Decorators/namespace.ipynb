```python
# local and gloal variable

a = 7 #=> global scope

def fun(): # local scope
    b = 9
    print(b)

fun()
print(a)
```

```python
# 2 scope of same name variable
a = 7 #=> global scope

def fun(): # local scope
    a = 9
    print(a)

fun()
print(a)
```

```python
# not in local but in gobal 
a = 7 #=> global scope

def fun(): # local scope
    print(a)

fun()
print(a)

# print ho raha hai , (local) mai nahi mill to enclose mai, {enclose} mai nahi to {global} mai
```

## Enclosing

```python
# usally seen in nested function 

def outer():
    def inner():
        print("Inner function")
    inner()
    print("Outer function")

outer()
print("Gobal main program")

# "Gobal main program"  => GLOBAL
# "Inner function" => LOCAL
# "Outer function" => ENCLOSING
```

```python
# clanging nonlocal vriable
def outer():
    a = 10
    def inner():
        nonlocal a
        a +=1
        print("Inner function",a)
    inner()
    print("Outer function",a)

outer()
print("Gobal main program")
```
