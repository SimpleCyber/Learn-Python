### Pandas - Series


```python
# it is an 1 dimensional column of an table , holding any data type => series
```

```python
# pandas us build upon NUMPY

# importing
import pandas as pd
import numpy as np
```

```python
# series form  a list

country = ['India','pak','usa','nepal','Butan']

# now we call a series class , and pass "country " in its constructor

pd.Series(country)

# series => consist of inex and values

# dtype: object ==> pandas "string" ko "object hi bolta hai"
```

```python
# integer

runs = [13,45,32,56,78,90,100]
pd.Series(runs)
```

```python
# Coustom Index using{Index attribute}
marks = [56,78,98, 101]
subject = ['english', 'maths','hidi','AMM']

# now creating
pd.Series(marks, index= subject)
```

```python
# setting a name  , so we can call it latter {using Name attribute}

marks = pd.Series(marks, index= subject, name= "Satyam ke marks")
marks
```

### Series from dict


```python
# same thing , custom indexing ,using dictionary

marks ={
"english" : 56,
"maths"  : 78,
"hidi"  :  98,
"AMM"  :  101
}

marks_series = pd.Series(marks, name = "Satyam ke marks")
marks_series  # it will result same output
```
