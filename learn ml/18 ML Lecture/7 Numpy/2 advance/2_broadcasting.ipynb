```python
# We work with array of different sizes, we see how smaller array work during the arithmatic operation  
```

```python
import numpy as np
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)


print(a)
print(b)

a+b
```

```python
a = np.arange(6).reshape(2,3)
b = np.arange(6,9)


print(a)
print(b)

a+b
```

```python
# examples

a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a)
print(b)

print(a+b)


```
