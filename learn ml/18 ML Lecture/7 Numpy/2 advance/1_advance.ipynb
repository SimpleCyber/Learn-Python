### Numpy Arrays vs Python Lists

```python
# speed
# list

a = [i for i in range(10000000)]
b = [i for i in range(10000000, 20000000)]

c = []
import time

start = time.time()

for i in range (10000000):
    c.append(a[i] + b[i])

print(time.time() - start)

# dynamic array , referential array , stores according to address
```

```python
# numpy  => uses c type of array, non dynamic 
import numpy as np

a = np.arange(10000000)
b = np.arange(10000000, 20000000)

start = time.time()
c = a + b
print(time.time() - start)
```

```python
# memory and convinent

a = [i for i in range(10000000)]

import sys

sys.getsizeof(a)
```

```python
a = np.arange(10000000 , dtype= np.int32)
sys.getsizeof(a)
```

```python
a = np.random.randint(1,100, 24).reshape(6,4)
a
```

```python
# find greater than 50

print(a > 50)
# the output is an boolean array

# boolean masking now we will , put this output on a , and were is true it will print elese it will no print

a[a>50]
```

```python
# find out even numbers

a % 2 == 0

a[a % 2 == 0]
```

```python
# graeter than 50 , and even

(a > 50) & (a % 2 == 0)

a[(a > 50) & (a % 2 == 0)]
```

```python
# numbers not divisibel by 7

a[~(a % 7 == 0)]
```
