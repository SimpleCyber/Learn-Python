```python
# sigmoid
def sigmoid(array):
    return 1/(1+np.exp(-(array)))

a = np.arange(10)

sigmoid(a)
```

```python
# mean squared error

# loss function => kin cheejo ko bhaden yea ghatana se bata hai ki error reduce hoata hai 

#1) is logistic Regression is mean squared error

actural = np.random.randint(1,50, 25)
prdicted = np.random.randint(1,50, 25)

print(actural)
print()
print(prdicted)

def mse(actural, prdicted):
    return np.mean((actural - prdicted)**2)


mse(actural , prdicted)

```
