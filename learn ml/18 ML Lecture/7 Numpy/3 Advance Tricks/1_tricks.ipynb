##### np.corrcoef ðŸŒŸ
###### eak bhadne se dusare pe kay effect padta hai
######  [-1 , 1]  lies between these , -1 inverse , 0 no relation , 1 direct relation

```python
salary = np.array([10000, 20000, 30000, 40000 , 45000])

experince = np.array([1,2,4,2,6])

np.corrcoef(salary , experince)


#             salary    experince
# salary        1.         0.74
# experince     0.74       1.

# 2 quantity 2x2 , if 3 then 3x3
```

##### set functions


```python
# np.union1d

m = np.array([1,2,3,4,5])
n = np.array([4,8,3,7,5])

np.union1d(m,n)
```

```python
# np.intersect1d

m = np.array([1,2,3,4,5])
n = np.array([4,8,3,7,5])

np.intersect1d(m,n)
```

```python
np.setdiff1d(m,n) # m mai hai per n mai nahi
```

```python
# remove common
np.setxor1d(m,n)
```

```python
# in  1 d => bo cheej hai ki nahi
np.in1d(m,1)
```
