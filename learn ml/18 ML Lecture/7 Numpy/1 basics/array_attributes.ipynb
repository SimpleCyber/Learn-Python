Array Functions

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100,3)

a1
```

```python
# max/min/sum/prod

np.max(a1)

np.sum(a1)
```

```python
# max/min/sum/prod  => row wise

# 0=> col , 1 => row

np.max(a1, axis=1)
```

```python
# mean , median , std , var

np.median(a1, axis= 0)
```

```python
# trignometric functions -> not using
np.sin(a1)
```

```python
# dot product , (3, 4 ),(4, 3) , 4=4 , then only dot product

a1 = np.arange(0,12).reshape(3,4)
a2 = np.arange(12,24).reshape(4,3)


np.dot(a1,a2)
```

```python
# log and exponants
np.log(a1)

np.exp(a1)
```

Stacking

```python
# horizontal stackup

a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)

a5
```

```python
np.hstack((a4, a5 , a4))
```

```python
# vertical stacking

np.vstack((a4 , a5))

# shape should be same
```

Splitting

```python
# hprizontal splitting  (  | ) , ooposite of stack
a6 = np.hstack((a4, a5 , a4))

```
