```python
class Point :
    def __init__(self, x , y):
        self.x_cord = x
        self.y_cord = y

    def __str__(self):
        return '<{},{}>'.format(self.x_cord , self.y_cord)
    
    def euclidean_distance (self , other):
        # distance formula  square_root((x2-x1)pow2 +(y2-y1)pow2)
        return ((self.x_cord - other.x_cord)**2 + (self.y_cord - other.y_cord)**2) **0.5
    

    def distance_from_origin(self):
        return self.euclidean_distance(Point(0,0))
    

class Line:
    def __init__(self ,A,B,C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return '{}x +{}y +{} = 0'.format(self.A,self.B,self.C)
    

    def point_on_line(line , point):
        if ((line.A * point.x_cord) +(line.B * point.y_cord) + line.C) == 0:
            return ("Point lies on line")
        
        else:
            return('Point not lies on line')


```

```python
l1 = Line(1,2,3)
print(l1)

p2 = Point(0 ,5)
print(l1.point_on_line(p2))
```
