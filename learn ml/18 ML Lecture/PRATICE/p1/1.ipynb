```python
class Atm:
    def __init__(self):
        self.pin = ''
        self.balance = 0
        print('Self address :', id(self))
        self.menu()

    def menu(self):
        user_input = input("""
        Hi how i can help u ? \n
        1. press 1 to create a pin \n
        2. press 2 to change a pin \n
        3. press 3 to check balance \n
        4. press 4 to Withdraw \n
        5. press anything to exit \n
        """) 
        match user_input:
            case "1":
                self.create_pin()
            case "2":
                self.change_pin()
            case "3":
                self.check_balance()
            case "4":
                self.withdraw_amt()
            case _:
                print('Thak u for Visiting Satyam Bank')

    
    def create_pin(self):
        user_pin = input('Enter your pin :')
        self.pin = user_pin

        user_balance = int((input('Enter Balance :')))

        self.balance = user_balance

        print("Pin Ban gaya tera")
        self.menu()

    def change_pin(self):
            old_pin = input("Enter old pin :")
            if self.pin == old_pin:
                 user_pin = input("Enter New pin:")
                 self.pin = user_pin
                 print('Pin Update ho gaya tera')
            else:
                 print("Sale choor ")
            
            self.menu()
        
    def check_balance(self):
        userpin = input("Enter youe PIN :")
        if userpin == self.pin:
            print("Current amount balance is :",self.balance)
        else:
            print("Sale chooor ")
        self.menu()

    def withdraw_amt(self):
        userpin = input("Enter youe PIN :")

        if userpin == self.pin:

            amount = int(input("Enter the withdraw amt :"))

            if amount <= self.balance:
                self.balance -= amount
            else:
                print("Graib ðŸ”")
            print("Current amount balance is :",self.balance)

        else:
            print("Sale chooor ")
        self.menu()



        
                
```
