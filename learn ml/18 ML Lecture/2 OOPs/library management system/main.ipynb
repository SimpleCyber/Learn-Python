```python
class Book:
    # constructor : initalize a new book object with basic details
    def __init__(self , title , author , genre , isbn):
        self.title = title   #public attribhte : title of book
        self.author = author
        self.genre = genre
        self.isbn = isbn
        self.is_availabe = True # public attribhte : check ststus of the book

    # method for update the abavilabity of the book
    def update_availability(self , status):
        self.is_availabe = status

    # method to get details of th book
    def get_details(self):
        return f"Title :{self.title}, Author :{self.author} , Genre :{self.genre}, Avaiability :{self.is_availabe}"
    

class Member:
    # Constructor: initializes a new Member object
    def __init__(self,name , member_id):
        self.name = name   # Public attribute: name of the member
        self.member_id = member_id
        self.borrowed_books = []     # Public attribute: list of borrowed books
    

    # method for a member to borruw a book
    def borrowed_books(self , book):
        if book.is_availabe:
            book.update_availability(False) # Update book's availability
            self.borrowed_books.append(book)    # Add book to member's borrowed list
            return f"{self.name} has borrowed {book.title}."
        else:
            return f"{book.title} is not availabe"
        
    
    # method to return a book
    def return_book(self , book):
        book.update_availability(True)
        self.borrowed_books.remove(book)
        return f"{self.name} has returned {book.title}"
    


# Now, let's create a simple example using these classes:

# creating  a book instance 
book1 = Book("Tp kill a Mockingbird", "Harper Lee", "Fiction", "12345")

# creating a member instance 
member1  = Member("Satyam yadav", "Moo1")

# memeber borrown a book
print(member1.borrowed_books(book1))
print(book1.get_details())   # Check book details to see if it's been borrowed

# Member returning a book
print(member1.return_book(book1))
print(book1.get_details())   # Check book details to see if it's available again


         
```
