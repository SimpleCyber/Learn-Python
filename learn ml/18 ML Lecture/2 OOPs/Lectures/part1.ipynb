```python
# our first datatype

class Fraction:
    # think first what we need

    # paramaterised constructor
    def __init__(self , x ,y):
        self.num = x
        self.den = y

    ######---------------------------------> 2nd magic method
        # if i will call print my object this magic code 
        # will be searche d in my python code 
        # and what the value it return it will be printed
        # output is according to me
    def __str__(self):
        return '{}/{}'.format(self.num , self.den)
    


    ######---------------------------------> 3nd magic method
    def __add__(self,other):
        # if in will try to add my 2 objects , 
        # it will be automatically executed

        # it willl search here for logic
        new_num = (self.num * other.den ) + (self.den * other.num)
        new_den = self.den * other.den

        return '{}/{}'.format(new_num , new_den)



    ######---------------------------------> 4nd magic method
    def __sub__(self,other):
        # if in will try to add my 2 objects , 
        # it will be automatically executed

        # it willl search here for logic
        new_num = (self.num * other.den ) - (self.den * other.num)
        new_den = self.den * other.den

        return '{}/{}'.format(new_num , new_den)
    

    ######---------------------------------> 5nd magic method
    def __mul__(self,other):
        # if in will try to add my 2 objects , 
        # it will be automatically executed

        # it willl search here for logic
        new_num = self.num * other.num 
        new_den = self.den * other.den

        return '{}/{}'.format(new_num , new_den)
    


        ######---------------------------------> 6nd magic method
    def __truediv__(self,other):
        # if in will try to add my 2 objects , 
        # it will be automatically executed

        # it willl search here for logic
        # div just reverse one side and multiply between them
        new_num = self.num * other.den 
        new_den = self.den * other.num

        return '{}/{}'.format(new_num , new_den)
    
    def conver_to_dec (self):
        return self.num / self.den
        


    


```

```python
frac1 = Fraction(3, 5)
frac2 = Fraction(3, 9)
```
