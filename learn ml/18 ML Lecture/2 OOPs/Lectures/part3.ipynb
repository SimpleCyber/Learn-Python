```python
# aggerigation

# EXAMPLE
# 2 class (costumer , address) like costumer has a address

# during aggregation private variables can not be accessed

# u can use getter and setter methods


class costumer:

    def __init__(self, name , gender , address):
        self.name = name 
        self.gender  = gender
        self.address = address

    
    # print address fetching data from second class(address)
    def print_address(self):
        print(self.address.get_city() , self.address.pincode , self.address.state)

    # edit the profile method
    def edit_profile(self , new_name , new_city , new_pin , new_state ):
        self.name = new_name
        self.address.set_city(new_city)
        self.address.pincode = new_pin
        self.address.state = new_state

class address:

    def __init__(self , city , pincode , state):
        # city is an private variable can be excessed within the class only
        self.__city = city
        self.pincode = pincode
        self.state = state
    
    # getter method to fetch the city 
    def get_city(self):
        return self.__city
    
    def set_city(self,new_city):
        self.__city = new_city



# objects of classes
add = address('Pune', '201211', 'Maharashtara')
cust = costumer('Satyam', 'Male', add )

cust.print_address()

cust.edit_profile("Rolli" ,'boisar', '401501', 'uttarpradesh')

cust.print_address()


```

```python
# CONCEPT 1 => agar child ke pass kud ka constructor naghi hai to parents ki values ko initalise karde ga aur parents ke class ke constructor ko execute jar de ga

# EXAMPLE

class Phone:
    def __init__(self , price , brand, camera):
        print("Inside the parent class constructor")
        self.price = price
        self.brand = brand
        self.camera = camera
    
    def buy(self):
        print("phone kharidne aaye hai")
        
# child vclass with no constructor
class SmartPhone(Phone):
    pass


realme = SmartPhone('10,000', 'Realmi' ,'60 mpxl')


realme.buy()
```

```python
# CONCEPT 2 => agar child ke pass kud ka constructor  hai to parents ka construcor kabi call nahi ho ga 

#aur parents ki koi be variables kabi nai bane ge 


# EXAMPLE

class Phone:
    def __init__(self , price , brand, camera):
        print("Inside the parent class constructor")
        self.price = price
        self.brand = brand
        self.camera = camera
    
    def buy(self):
        print("phone kharidne aaye hai")
        

# clild calss with construcctor
class SmartPhone(Phone):
    def __init__(self, os ,ram):
        print("insde the child calss construcor")
        self.os  = os
        self.ram = ram


realme = SmartPhone('Android', '4gb')
realme.buy()
```

Super Keyword

```python
 # EXAMPLE

class Phone:
    def __init__(self , price , brand, camera):
        print("Inside the parent class constructor")
        self.price = price
        self.brand = brand
        self.camera = camera
    
    def buy(self):
        print("phone kharidne aaye hai")
        

# clild calss with construcctor
class SmartPhone(Phone):
    
    def buy(self):
        print("SmartPhone kharidne aaye hai")

        # using super ðŸ˜ƒðŸ˜ƒðŸ˜ƒ
        super().buy()


realme = SmartPhone(10000 , 'Realme' ,60)
realme.buy()
```
