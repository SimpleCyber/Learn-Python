```python
from abc import ABC,abstractmethod

# RULES 
# 1)  from abc import ABC,abstractmethod
# 2)  class BankApp(ABC):  compulsory
# 3)  @abstractmethod

# then only a class is an abstract method


class BankApp(ABC):
    def database(self):
        print("Connected to database")
    
    @abstractmethod
    def sequrity(self):
        pass

    @abstractmethod
    def dispaly(self):
        pass


```

```python
class MobileApp(BankApp):

    def mobile_login(self):
        print("Mobile login")


    # jabtak [sequrity(self)] iska code nahi ho ga tab tak object nahi bane ga
    def sequrity(self):
        print("MobileApp security") 


    # through these we can restrict the child class with some restriction
        
    def dispaly(self):
        print("dispaly")
    

        


    
```

```python
obj  = MobileApp()
obj.database()
obj.sequrity()
```
